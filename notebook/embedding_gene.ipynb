```python
path = Path("../output/metrics/20251022_All_BindingDB_prediction_analysis_by_class_level_6.csv")
df = pd.read_csv(path)
df.head()
```

```python
df.query("Gene	 == 'RPS6KA3'")
```

```python
# Alternative: Group by embedding method
fig, ax = plt.subplots(figsize=(12, 8))

# Pivot with embedding methods as index
plot_data_alt = df.pivot_table(
    index='Embedding', 
    columns='Disease_Name', 
    values='pearson',
    aggfunc='first'
)

plot_data_alt.plot(kind='bar', ax=ax, width=0.8)

ax.set_title('Performance Comparison Across Diseases by Embedding Method', 
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Embedding Method', fontsize=12, fontweight='bold')
ax.set_ylabel('Pearson Correlation', fontsize=12, fontweight='bold')

plt.xticks(rotation=0)  # Keep embedding method names horizontal
ax.legend(title='Disease', title_fontsize=12, fontsize=9, 
          bbox_to_anchor=(1.05, 1), loc='upper left')

ax.grid(axis='y', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()
```
